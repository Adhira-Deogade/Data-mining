## Creating labels and annotation


```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sb
```

```python
%matplotlib inline 
# instead of opening in other window
rcParams['figure.figsize']=5,4
# 5 inches wide and 4 inches high
sb.set_style('whitegrid')
```

##### Functional method

```python
x = range(1,10)
y  = [1,2,3,4,0.5,4,3,2,1]
plt.bar(x,y)

plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
# labelling pie chart

z = [1, 2, 3, 4, 0.5]
veh_type = ['bicycle','motorbike','car','van','stroller']
plt.pie(z, labels=veh_type)
plt.show()
```

```python
# using object orienred method
address = '/Users/Adhira/Desktop/Python ML/Ex_Files_Python_Data_Science_EssT/Exercise Files/Ch02/02_01/mtcars.csv'
cars = pd.read_csv(address)

cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']

mpg = cars.mpg

fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])
mpg.plot()

ax.set_xticks(range(32))

ax.set_xticklabels(cars.car_names,rotation=60,fontsize='medium')

ax.set_title('Miles per gallon of cars in mtcars')

ax.set_xlabel('Car names')
ax.set_ylabel('Miles per gallon')
```

```python
plt.pie(z)

# Adding legend

plt.legend(veh_type, loc = 'best')
plt.show()
```

```python
# using object oriented method

fig = plt.figure()

ax =fig.add_axes([.1,.1,1,1])

mpg.plot()

ax.set_xticks(range(32))

ax.set_xticklabels(cars.car_names,rotation=60,fontsize='medium')

ax.set_title('Miles per gallon of cars in mtcars')

ax.set_xlabel('Car names')
ax.set_ylabel('Miles per gallon')

ax.legend(loc ='best')
```
